```python
#Fonction cree_reseau demandée a la question 1

def cree_reseau(tab) :
    reseau = {}
    i = 0
    while i < len(tab)/2:
        a = tab[2*i]
        b = tab[2*i+1]
        if a not in reseau :
            reseau[a] = []
        if b not in reseau :
            reseau[b] = []
        if b not in reseau[a] :
            reseau[a].append(b)
        if a not in reseau[b] :
            reseau[b].append(a)
        i += 1
    return reseau


amis = ["Alice", "Bob", "Alice", "Dan", "Bob", "Carl", "Bob", "Dan","Alice", "Bob" ]

cree_reseau(amis)
```

```python
#Question 2

"""
La fonction "cree_reseau" traverse 1 seule fois un tableau d'amis tandis que 
la fonction "dico_reseau" (celle de la correction de la SAE en tout cas) ci
dessous utilises des fonctions intermédiaires "personne" (cree un tableau de 
toute les personnes présentes dans le grand tableau représantant les 
interactions entre les personnes) "ses_amis" qui regroupe tout les amis de 
"prenom" passé en paramètre dans le tableau représentant les interactions
entre les personnes. La fonction parcours le grand tableau pour chaque clé 
d'un dictionnnaire représentant une personne afin de trouver ses amis et les 
ajouter a un tableau associé a la clé portant le nom de la personne

Complexité de "dico_reseau" : A chaque itération de la boucle de "dico_reseau",
la fonction appelle "ses_amis", qui utilise une boucle également, alors la
complexité est quadratique O(n²)

Complexité de "cree_reseau" : A chaque itération de la boucle de la fonction, 
on parcours le dictionnaire initialisé "réseau" avec la fonction "not in" par exemple,
donc cela fonctionne comme une boucle dans une autre, alors la complexité est quadratique O(n²)

En théorie, ces deux fonctions ont la même complexité asymptotique.
"""


from time import *


def personnes(amis):
    """
    retourne le tableau des différentes personnes du tableau amis
    """
    personnes=[]
    i=0
    while i<len(amis):
        if amis[i] not in personnes:
            personnes.append(amis[i])
        i+=1
    return personnes

def ses_amis(amis, prenom):
    """
    Retourne le tableau des amis de prenom
    """
    ses_amis=[]
    i=0
    while i<len(amis)/2:
        if amis[2*i]== prenom :
            ses_amis.append(amis[2*i+1])
        elif amis[2*i+1]==prenom :
            ses_amis.append(amis[2*i])
        i+=1
    return ses_amis


#version non optimisée 
def dico_reseau(amis):
    reseau={}
    # membres du réseau 
    pers=personnes(amis)
    # construction du dictionnaire
    i=0
    while i<len(pers):
        reseau[pers[i]]=ses_amis(amis,pers[i])
        i+=1
    return reseau




#--------------------------------------------------------------#


# Fonction pour mesurer le temps moyen pour créer un dictionnaire avec dico_reseau
def mesure_dico_reseau(amis): 
    nb_mesures = 500  # On fait nb_mesures de la vitesse de creation d'un dico
    tic = time() # top départ
    i = 0
    while i < nb_mesures:
        dico_reseau(amis)
        i+=1   
    tac = time()    # arrêt du chronomètre
    
    # On fait la moyenne des temps d'exécutions (ms) des nb_mesures exécutions et on arrondit au millième
    return round(1000*(tac-tic) / nb_mesures, 3)

# Fonction pour mesurer le temps moyen pour créer un dictionnaire avec cree_reseau, on procède comme précédemment
def mesure_cree_reseau(amis):
    nb_mesures = 500    
    tic = time()
    i = 0
    while i < nb_mesures:
        cree_reseau(amis)
        i+=1
    tac = time()
    return round(1000*(tac-tic) / nb_mesures, 3)

print("temps d'execution de la fonction dico_reseau :", mesure_dico_reseau(amis))
print("temps d'execution de la fonction cree_reseau :", mesure_cree_reseau(amis))


"""
On constate alors que cree_reseau est plus rapide que dico_reseau avec ces test pratiques

"""
```

```python
#Question 11 :

"""
La fonction comu_dans_amis est conçue pour trouver la plus grande communauté possible autour 
d’une personne donnée ou d’un groupe restreint spécifié. Elle analyse localement les connexions 
d’une personne et vérifie quelles relations amicales permettent de former une communauté où tout 
le monde est connecté. En revanche, comu_dans_reseau a une portée globale : elle cherche la plus 
grande communauté dans l’ensemble du réseau, en explorant toutes les relations entre les membres.

Complexité de "comu_dans_amis" : A chaque itération de la boucle for de "comu_dans_amis",
la fonction appelle "sont_amis_de", qui utilise une boucle également, alors la
complexité est quadratique O(n²)

Complexité de "comu_dans_reseau" : La fonction utilise le tri à bulle, qui est de complexité quadratique O(n²)

En théorie, ces deux fonctions ont la même complexité asymptotique.
"""





```

```python
#Fonctions nécéssaires au fonctionnement de "comu_dans_amis" et "comu_dans_reseau"


def liste_personnes(reseau):
    return list(reseau.keys())  # Retourne les clés du dictionnaire, qui sont les noms des personnes



def sont_amis(personne1, personne2, reseau):
    if personne1 in reseau:  # Vérifie si personne1 est dans le réseau
        if personne2 in reseau[personne1]:  # Vérifie si personne2 est dans la liste des amis de personne1
            return True  # Elles sont amies
    return False  # Elles ne sont pas amies


def sont_amis_de(personne, groupe, reseau):
    for membre in groupe:  # Parcourt tous les membres du groupe
        if not sont_amis(personne, membre, reseau):  # Si la personne n'est pas amie avec un membre, retourne False
            return False
    return True  # Retourne True si la personne est amie avec tout le groupe

def comu(groupe, reseau):
	i = 0
	commu = [] # La communauté commence vide
	while i < len(groupe) :  #On regarde chaque personne du groupe
		personne = groupe[i]
		if sont_amis_de(personne,commu,reseau) : # si la personne est pote avec un membre
			commu.append(personne) # Alors elle peut rejoindre
		i += 1
    
	return commu 

def tri_popu(groupe,reseau):
    grp = groupe.copy()
    i = 0
    while i < len(grp) - 1 :
        j = 0
        while j < len(grp) - i - 1 :
            if len(reseau[grp[j]]) < len(reseau[grp[j+1]]) :
                grp[j] , grp[j+1] = grp[j+1] , grp[j]
            j += 1
        i += 1
    return grp
    
```

```python
def comu_dans_amis(personne, reseau):
	amis = tri_popu(reseau[personne],reseau)
	groupe = [personne]  # Initialise le groupe avec la personne donnée
	for ami in amis:  # Parcourt les amis de cette personne
		if sont_amis_de(ami, groupe, reseau):  # Vérifie si l'ami est compatible avec le groupe
			groupe.append(ami)  # Ajoute l'ami au groupe si la condition est remplie
	return groupe  # Retourne le groupe formé

def comu_dans_reseau(reseau):
	i = 0
	personnes = liste_personnes(reseau)  # stockage du tableau avec toutes les personnes du reseau
	while i < len(personnes)-1 :
		j = 0
		while j < len(personnes)-i-1 :
			if len(reseau[personnes[i]]) < len(reseau[personnes[i + 1]]) : #Comparaison de leur nombre d'amis
				personnes[j] , personnes[j+1] = personnes[j+1] , personnes[j]
			j += 1
		i += 1
	tab = comu(personnes, reseau)
	return tab


personne = "Alice"
```

```python
# Fonction pour mesurer le temps moyen pour créer un dictionnaire avec dico_reseau
def mesure_comu_dans_amis(personne, reseau): 
    nb_mesures = 500  # On fait nb_mesures de la vitesse de creation d'un dico
    tic = time() # top départ
    i = 0
    while i < nb_mesures:
        comu_dans_amis(personne,reseau)
        i+=1   
    tac = time()    # arrêt du chronomètre
    
    # On fait la moyenne des temps d'exécutions (ms) des nb_mesures exécutions et on arrondit au millième
    return round(1000*(tac-tic) / nb_mesures, 3)

# Fonction pour mesurer le temps moyen pour créer un dictionnaire avec cree_reseau, on procède comme précédemment
def mesure_comu_dans_reseau(reseau):
    nb_mesures = 500    
    tic = time()
    i = 0
    while i < nb_mesures:
        comu_dans_reseau(reseau)
        i+=1
    tac = time()
    return round(1000*(tac-tic) / nb_mesures, 3)

print("temps d'execution de la fonction comu_dans_reseau :", mesure_comu_dans_reseau(reseau))
print("temps d'execution de la fonction comu_dans_amis :", mesure_comu_dans_amis(personne, reseau))

"""
On constate que comu_dans_amis est plus rapide que comu_dans_reseau avec ces tests pratiques

"""
```
